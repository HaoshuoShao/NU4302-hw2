## CONDITION
$$
u = 5m/s \\
\frac{\partial \phi}{\partial t}+ u\frac{\partial \phi}{\partial x} = 0\\
\phi = 0\\
t = 0\\
x = 0, \phi =1, 0\le t \le 0.02s\\
x = 0, \phi =0, t > 0.02s\\
$$



```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Define the dimensions of the 2-D matrix
spatial_dim = 50
temporal_dim = 100
num_frames = 100

# Generate random data for the 2-D matrix
A = np.random.rand(spatial_dim, temporal_dim)

# Create a figure and axis
fig, ax = plt.subplots()

# Initialize the plot with an empty image
img = ax.imshow(A, cmap='viridis', aspect='auto', origin='lower', animated=True)

# Define the initialization function
def init():
    img.set_data(np.zeros_like(A))
    return img,

# Define the update function
def update(frame):
    # Simulate the evolution of the data
    A[:, :-1] = A[:, 1:]
    A[:, -1] = np.random.rand(spatial_dim)  # Add new random data at the last temporal position
    img.set_data(A)
    return img,

# Create the animation
ani = FuncAnimation(fig, update, frames=num_frames, init_func=init, blit=True)

# Display the animation in Jupyter Notebook
HTML(ani.to_jshtml())
```
