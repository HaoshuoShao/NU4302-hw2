## CONDITION
$$
u = 5m/s \\
\frac{\partial \phi}{\partial t}+ u\frac{\partial \phi}{\partial x} = 0\\
\phi = 0\\
t = 0\\
x = 0, \phi =1, 0\le t \le 0.02s\\
x = 0, \phi =0, t > 0.02s\\
$$



```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
dx = 0.0005
dt = 0.0001
length = 1
x_num = int(length/dx)
xt_ratio = dx/dt
time = 0.1
t_num = int(time/dt)
phi = np.zeros((t_num, x_num))

#initialization
for index in range(0, int(t_num/5)):
    phi[index, 0] = 1

A = np.zeros((x_num, x_num))
b = np.zeros(x_num)
```

```python
#LUD
A[0, 0] = 10 + xt_ratio
A[1, 0] = -25/2
A[1, 1] = 15/2 + xt_ratio
for x in range(2, x_num):
    A[x, x-2] = 5/2
    A[x, x-1] = -10
    A[x, x] = 15/2 + xt_ratio

A_inverse = np.linalg.inv(A)

def get_b(phi_prev, xt_ratio, phi_0):
    b = phi_prev*xt_ratio
    b[0] += 10*phi_0
    b[1] -= 5*phi_0
    return b
```

```python
#CACULATE
for index in range(1, t_num):
    if index <= int(t_num/5):
        phi_0 = 1
    else:
        phi_0 = 0
    b = get_b(phi[index], xt_ratio, phi_0)
    x = A_inverse.dot(b)

    phi[index] = x
print(phi[:50, :20])
```

```python
num_time_steps = t_num
num_spatial_points = x_num
phi_matrix = phi

# 创建一个画布
fig, ax = plt.subplots()

# 创建一个空白的线图
line, = ax.plot([], [])
plt.xlim((0, 1))
plt.ylim((-0.1, 1.1))


# 定义初始化函数
def init():
    line.set_data([], [])
    return line,

# 定义更新函数，用于生成 phi-x 曲线
def update(frame):
    phi_x = phi_matrix[:, frame]  # 获取当前时间步的 phi-x 数据
    x = np.arange(len(phi_x))  # x 轴是空间位置
    line.set_data(x, phi_x)
    return line,

# 创建动画
ani = FuncAnimation(fig, update, frames=num_time_steps, init_func=init, blit=True)

# 在 Jupyter Notebook 中显示动画
HTML(ani.to_jshtml())
```
